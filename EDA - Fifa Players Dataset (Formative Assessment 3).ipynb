```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file into a DataFrame
df = pd.read_csv('E:\\DSML\\Assignments\\May30\\fifadata.csv')
# Check the structure of the DataFrame
#print(df.head())
#1.Which country has the most number of players (score :1)
most_players_country = df['Nationality'].value_counts().idxmax()
num_players_most_country = df['Nationality'].value_counts().max()
print(f"Insights:\nThe country with the most number of players is {most_players_country} with {num_players_most_country} players.")
```

```python
#2.Plot a bar chart of 5 top countries with the most number of players. (score :1)
# Group the data by country and count the number of players in each country
country_counts = df['Nationality'].value_counts()
# Get the top 5 countries with the most number of players
top_countries = country_counts.head(5)
# Plot the bar chart
plt.figure(figsize=(10, 4))
top_countries.plot(kind='bar', color='red')
plt.title('Top 5 Countries with the Most Number of Players')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.xticks(rotation=15)  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.show()
print("Insights :\n",top_countries, " are 5 top countries with the most number of players")
```

```python
#3.Which player has the highest salary? (score :1)
# Find the player with the highest salary
player_highest_salary = df.loc[df['Wage'].idxmax()]
# Display the player with the highest salary
print("Player with the highest salary:")
print(player_highest_salary[['Name', 'Wage']])
print("Insights :\n",player_highest_salary[['Name']],"has highest salary")
```

```python
#4.Plot a histogram to get the salary range of the players.
# Convert salary column to numeric (remove currency symbols, commas, etc.)
#df['Wage'] = df['Wage'].replace('[\â‚¬,]', '', regex=True).astype(float)
# Plot histogram
plt.figure(figsize=(9, 5))
plt.hist(df['Wage'], bins=20, color='skyblue', edgecolor='black')
plt.title('Salary Range of Players')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.tight_layout()
plt.show()
print("Insights :\nHighest frequecy is above 8000")
```

```python
#5.Who is the tallest player in the fifa?
ht=df.sort_values(by=["Height"], ascending=False)
print("Insights :")
ht[['Name','Height']].head(1)
```

```python
#6.Which club has the most number of players?
most_players_club = df['Club'].value_counts().idxmax()
num_players_most_club = df['Club'].value_counts().max()
print(f"Insights :\nThe club with the most number of players is {most_players_club} with {num_players_most_club} players.")
```

```python
#7.Which foot is most preferred by the players?Draw a bar chart for preferred foot
# Count the number of players preferring each foot
foot_counts = df['Preferred Foot'].value_counts()
# Plot the bar chart
plt.figure(figsize=(8, 6))
foot_counts.plot(kind='bar', color='skyblue')
plt.title('Preferred Foot of Players')
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.xticks(rotation=0)  # Rotate x-axis labels if needed
plt.tight_layout()
plt.show()
print("Insights :\n right foot is most preferred by the players")
```
